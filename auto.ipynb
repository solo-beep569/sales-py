```python
#importing libraries and loading data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
df = pd.read_csv("./data/auto.csv")
df
```

```python
#exploring the data
df.columns
df.info()
df.describe(include='all')
df.isnull().sum()
```

```python
#renaming columns
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df.columns = headers
df.head()
df
```

```python
#replace missing values represented as "?" with NaN
df.replace("?", np.nan, inplace=True)
df.head()
```

```python
#checking missing data
missing_data = df.isnull()
missing_data
```

```python
#checking missing data
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

```python
#dealing with missing data
avrg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
df["normalized-losses"].replace(np.nan, avrg_norm_loss, inplace=True)
```

```python
#replacing bore missing vallues with its mean
avrg_bore = df["bore"].astype("float").mean(axis=0)
df["bore"].replace(np.nan, avrg_bore, inplace=True)
```

```python
#replacing stroke missing vallues with its mean
avrg_stroke = df["stroke"].astype("float").mean(axis=0)
df["stroke"].replace(np.nan, avrg_stroke, inplace=True)
```

```python
#replacing horsepower missing vallues with its mean
avrg_horsepower = df["horsepower"].astype("float").mean(axis=0)
df["horsepower"].replace(np.nan, avrg_horsepower, inplace=True)
```

```python
#replacing peak-rpm missing vallues with its mean
avrg_peakrpm = df["peak-rpm"].astype("float").mean(axis=0)
df["peak-rpm"].replace(np.nan, avrg_peakrpm, inplace=True)
```

```python
#replacing num-of-doors missing vallues with its frequency
df["num-of-doors"].value_counts().idxmax()
df["num-of-doors"].replace(np.nan, "four", inplace=True)
```

```python
#dropping an empty row for price
df.dropna(subset=["price"], axis=0, inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

```python
#checking the various data types
df.dtypes
```

```python
#
df.describe(include="all")
```

```python
#changing city-mpg to city-L/100km
df['city-L/100km'] = 235/df["city-mpg"]
df.head()
```

```python
#changing highway-mpg to highway-L/100km
df["highway-L/100km"] = 235/df["highway-mpg"]
df.head()
```

```python
#normalizing or scaling length, width and height
df["length"] = df["length"]/df["length"].max()
df["width"] = df["width"]/df["width"].max()
df["height"] = df["height"]/df["height"].max()
df.head()
```

```python
#plot a histogram of horsepower
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
df["horsepower"] = df["horsepower"].astype(int, copy=True)

plt.pyplot.hist(df["horsepower"])
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
#creating a low, medium and high group for the bins
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
group_names = ['Low', 'Medium', 'High']
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head(20)
```

```python
#counting the number of vehicles in each bin
df['horsepower-binned'].value_counts().idxmax
```

```python
#plotting the distribution of each bin
pyplot.bar(group_names, df['horsepower-binned'].value_counts())
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")  
plt.pyplot.title("horsepower bins")
```

```python
plt.pyplot.hist(df['horsepower'], bins = 3)
plt.pyplot.xlabel("horsepower") 
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
#creating dummy variables for fuel type
dummy_variable_1 = pd.get_dummies(df["fuel-type"])
dummy_variable_1.head()
```

```python
#renaming the columns
dummy_variable_1.rename(columns={'gas':'fuel-type-gas', 'diesel':'fuel-type-diesel'}, inplace=True)
dummy_variable_1.head()
```

```python
#concatenating the dummy variables to the original dataset and dropping the fuel-type column
df = pd.concat([df, dummy_variable_1], axis=1)
df.drop("fuel-type", axis = 1, inplace=True)
df.head(20)
```

```python
#Creating dummy variables for aspiration
dummy_variable_2 = pd.get_dummies(df["aspiration"])
dummy_variable_2.rename(columns={'std':'aspiration-std', 'turbo':'aspiration-turbo'}, inplace=True)
dummy_variable_2.head()

df = pd.concat([df, dummy_variable_2], axis=1)
df.drop("aspiration", axis = 1, inplace=True)   
df.head()
```
