Analyzing Data


```python
import pandas as pd 
pd.options.display.max_rows = 9999
data = pd.read_csv('./data/data.csv')
print(data)
```

```python
drop_data = data.dropna()
print(drop_data)
drop_data.corr()
```

```python
#healthy people fall below the average number of Calories consumed
Average_Calories = drop_data['Calories'].mean()
healthy_people = drop_data[drop_data['Calories'] < Average_Calories]
print(healthy_people)
```

```python
#making a table showing healthy people
import pandas as pd
healthy_people = pd.Series(drop_data.Calories < Average_Calories)

healthy_people_data = pd.concat([drop_data, healthy_people.rename('Healthy')], axis=1)
print(healthy_people_data)
```

```python
healthy_people_data.groupby(healthy_people_data.Healthy).mean()
```

```python
#sorting 
```

charting and plot


```python
from matplotlib import pyplot as plt
plt.bar(x=drop_data.Pulse, height=drop_data.Calories, color='orange', width=2)

plt.title('Pulse vs Calories')
plt.xlabel('Pulse')
plt.ylabel('Calories')
plt.show()

healthy_count = healthy_people_data.Healthy.value_counts() 
plt.pie(healthy_count, labels=['Unhealthy', 'Healthy'], autopct='%1.1f%%', startangle=90, colors=['red', 'green'])
plt.title('Proportion of Healthy vs Unhealthy People')
plt.show() 
```

```python
mean_val = healthy_people_data.Healthy.mean()
print(mean_val)
mode_val = healthy_people_data.Healthy.mode()[0]
print(mode_val)
median_val = healthy_people_data.Healthy.median()
print(median_val)
max_val = healthy_people_data.Healthy.max()
print(max_val)  
min_val = healthy_people_data.Healthy.min()
print(min_val)  
 
```
