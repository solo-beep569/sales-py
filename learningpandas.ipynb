```python
import matplotlib.pyplot as plt
```

```python


fig, ax = plt.subplots(figsize=(10, 25))
ax.axis('off')

boxes = [
    (0.5, 0.95, "Start Podcast Project"),
    (0.5, 0.88, "Project Planning & Goals"),
    (0.5, 0.81, "Concept & Branding"),
    (0.5, 0.74, "Concept Approved?"),
    (0.5, 0.66, "Content Planning & Research"),
    (0.5, 0.59, "Pre-Production Setup"),
    (0.5, 0.52, "Equipment & Audio OK?"),
    (0.5, 0.44, "Recording"),
    (0.5, 0.37, "Editing"),
    (0.5, 0.30, "Audio Quality Acceptable?"),
    (0.5, 0.22, "Publish & Distribute"),
    (0.5, 0.15, "Marketing & Promotion"),
    (0.5, 0.08, "Analytics & Improvement"),
]

plt.title("Work Breakdown Structure of a Podcast\n By\n Solomon Kofi Fiaku")

for x, y, text in boxes:
    ax.text(x, y, text, ha='center', va='center', bbox=dict(boxstyle="round,pad=0.9"))

for i in range(len(boxes) - 1):
    ax.annotate("", xy=(0.5, boxes[i+1][1]+0.03), xytext=(0.5, boxes[i][1]-0.03),
                arrowprops=dict(arrowstyle="->"))


path = "./dataguy/podcast_flowchart (1).png"
plt.savefig(path, bbox_inches='tight', dpi=200)
#plt.close()
plt.show()

path
```
